# Desenvolva um código que simule uma eleição com três candidatos.
- candidato_X => 889
- candidato_Y => 847
- candidato_Z => 515
- branco => 0

O código deve perguntar se deseja finalizar a votação depois do voto. Caso o número do voto não corresponda a nenhum candidato ou seja branco, ele deve ser tratado como nulo. Se for inserido um texto ao invés de número, o código deve retornar uma mensagem para votar novamente.
Quando a votação for finalizada, o código deverá mostrar o vencedor, aquele com o maior número de votos e, também, a quantidade de votos de cada candidato, os brancos e nulos 

Simular a eleição de 3 candidatos (Candidato_X,Candidato_Y,Candidato_Z,votos_brancos)
codigo de cada candidato:
- candidato_X => 889
- candidato_Y => 847
- candidato_Z => 515
- branco => 0
Caso o número do voto não corresponda a nenhum candidato ou seja branco, ele deve ser tratado como nulo
Se for inserido um texto ao invés de número, o código deve retornar uma mensagem para votar novamente.
O código deve perguntar se deseja finalizar a votação depois do voto
Quando a votação for finalizada, o código deverá mostrar:
-O vencedor com os seus respectivos votos
-A quantidade de votos de cada candidato
_Os votos brancos e nulos.

```python
print("INICIANDO VOTAÇÃO")
print('='*40)
print("\nlista opções com seu respectivos números:")
print("candidato_X => digite: 889")
print("candidato_Y => digite: 847")
print("candidato_Z => digite: 515")
print("Branco      => digite: 0")
print('='*40)


import enum
class Candidato(enum.Enum):
    candidato_x =889
    candidato_y =847
    candidato_z=515
    branco = 0
votar = True

X = 0
Y = 0
Z = 0
Vb= 0
Vn = 0

while votar:
    votoInvalido = True
    voto = 0
    while votoInvalido:
        try:
            voto = int(input("Informe o número do seu candidado:"))
            votoInvalido = False
        except:
            print("Informe um valor válido, Vote novamente.")
    if X>Y:
        if Y>=Z:candi = "X"
    if Y>X:
        if X>=Z:candi = "Y"
    if Z>X:
        if X>=Y:candi = "Z"
            
    if Candidato.candidato_x.value==voto:
        X+=1
    elif Candidato.candidato_y.value ==voto:
        Y+=1
    elif Candidato.candidato_z.value == voto:
        Z+=1
    elif Candidato.branco.value==voto:
        Vb+=1
    else:
        Vb+=1
    opcao = int(input('Desaja continuar votando? (1-sim e 2-não):'))
    if opcao ==1:
        votar=True      
    elif opcao ==2:
        votar=False
                  
print("="*40)
print('\nO candidato(a) mais votado(a) foi: {}'.format(candi))
print("\nVotos para o(a) candidato(a) {}: {}".format(Candidato.candidato_x.name,X))
print("Votos para o(a) candidato(a) {}: {}".format(Candidato.candidato_y.name,Y))
print("Votos para o(a) candidato(a) {}: {}".format(Candidato.candidato_z.name,Z))
print("\nVotos nulos: {}".format(Vn))
print("Votos brancos: {}".format(Vb))

   

```
